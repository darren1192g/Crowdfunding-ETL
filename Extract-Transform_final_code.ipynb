```python
import pandas as pd
import json
import re
import numpy as np
```

```python
# Read the data into a Pandas DataFrame
bi = pd.read_csv("backer_info.csv")
# Get the sheet names.
bi
```

```python
bi['backer_id'] = bi['backer_info'].str.extract(r'([a-z]+\d{3})')
bi.head()
```

```python
bi['backer_id'] = bi['backer_info'].str.extract(r'([a-z]+\d{3})')
bi.head(10)
```

```python
bi['cf_id'] = bi['backer_info'].str.extract(r'\s(\d+\d)')
bi.head()
```

```python
bi['name'] = bi['backer_info'].str.extract(r'([^nameil\s+][A-Za-z]+\s+[A-Za-z]+)').apply(lambda s:s.str.replace('"', ""))
bi.head()      
```

```python
bi['email'] = bi['backer_info'].str.extract(r'(\S+@\S+)').apply(lambda s:s.str.replace('"', ""))
bi.head()
```

```python
bi_df = bi[['backer_id','cf_id', 'name', 'email',]]
bi_df.head()
```

```python
bi_df.to_csv('backer_info.csv', encoding='utf-8')
```

```python
# Create a "first"name" and "last_name" column with the first and last names from the "name" column. 
bi_df[["first_name","last_name"]] = bi_df["name"].str.split(' ', n=1, expand=True)

# Drop the name column. 
bi_df_clean = bi_df.drop(['name'], axis=1)
bi_df_clean.head(10)
```

```python
bi_df_clean.info()
```

```python
bi_df_clean[['cf_id']] = bi_df_clean[['cf_id']].astype(int)
bi_df_clean.info()
```

```python
bi_df_clean = bi_df_clean[['backer_id', 'cf_id', 'first_name', 'last_name', 'email']]
bi_df_clean.head(10)
```

```python
bi_df_clean.to_csv('backers.csv', encoding='utf-8')
```
