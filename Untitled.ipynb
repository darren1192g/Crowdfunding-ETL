```python
import pandas as pd
import json
import re
import numpy as np
```

```python
# Read the data into a Pandas DataFrame
bi = pd.read_csv("backer_info.csv", header= [0, 1])
# Get the sheet names.
bi
```

```python
bi['backer_id'] = bi['backer_info'].str.extract(r'([^banker_idil\s+][\d])')
bi.head(10)
```

```python
bi['cf_id'] = bi['backer_info'].str.extract(r'([^cf_id\s+][\d])')
bi.head()
```

```python
bi['name'] = bi['backer_info'].str.extract(r'([^nameil\s+][A-Za-z]+\s+[A-Za-z]+)').apply(lambda s:s.str.replace('"', ""))
bi.head()      
```

```python
bi['email'] = bi['backer_info'].str.extract(r'(\S+@\S+)').apply(lambda s:s.str.replace('"', ""))
bi.head()
```

```python
bi_df = bi[['backer_id', 'name', 'email', 'cf_id']]
bi_df.head()
```
